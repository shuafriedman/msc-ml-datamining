
# Regularization "limits" the model optimization and is useful in noise



```python

import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model

X_train = np.c_[.5, 1].T
y_train = [.5, 1]
X_test = np.c_[0, 2].T

np.random.seed(0)

classifiers = dict(ols=linear_model.LinearRegression(),
                   ridge=linear_model.Ridge(alpha=.1))

for name, clf in classifiers.items():
    fig, ax = plt.subplots(figsize=(4, 3))

    #grey - noisy models
    for _ in range(6):
        this_X = .1 * np.random.normal(size=(2, 1)) + X_train
        clf.fit(this_X, y_train)

        ax.plot(X_test, clf.predict(X_test), color='gray')
        ax.scatter(this_X, y_train, s=3, c='gray', marker='o', zorder=10)

    #blue - clean model
    clf.fit(X_train, y_train)
    ax.plot(X_test, clf.predict(X_test), linewidth=2, color='blue')
    ax.scatter(X_train, y_train, s=30, c='red', marker='+', zorder=10)

    ax.set_title(name)
    ax.set_xlim(0, 2)
    ax.set_ylim((0, 1.6))
    ax.set_xlabel('X')
    ax.set_ylabel('y')

    fig.tight_layout()

plt.show()

#without reg. models try to get the optimum weights - more variance
#reg. models predictions are biased but with less variance
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Mar  8 11:53:08 2020

@author: yosi
"""

import numpy as np
import pandas as pd
#import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression,Ridge

table = pd.read_csv('C:\\WinPython-3.5\\notebooks\\docs\\weight-height.csv')
height = table['Height'][table['Gender'] == 'Male'].values
weight = table['Weight'][table['Gender'] == 'Male'].values

X = (weight-np.mean(weight))/np.std(weight)
y = height
```

```python
#Gradient descent with or without L2 regularization
def GD_reg(X,Y,alpha):#L2
    # Initialization
    W = 0
    b = 0
    L = 0.5 # The learning Rate
    epochs = 100  # The number of iterations to perform gradient descent

    n = float(len(X)) # Number of elements in X

    b_ = []
    W_ = []
    # Performing Gradient Descent
    for i in range(epochs):
        Y_pred = W*X + b  # The current predicted value of Y
        D_W = (-2/n) * sum(X * (Y - Y_pred)) + alpha*W # Derivative wrt W. Note the regularization (L2) term
        D_b = (-2/n) * sum(Y - Y_pred)  # Derivative wrt b. We don't apply regularization to intercept term
        W = W - L * D_W  # Update W
        b = b - L * D_b  # Update b
        W_.append(W)
        b_.append(b)
    return W_,b_
```

## Regularized models don't converge to optimum weights and lowest error

```python
#split into train and validation partitions
X_train = X[:2500];y_train = y[:2500]
X_val = X[2500:];y_val = y[2500:]

#sort the partitions
idx = np.argsort(X_train)
X_train = X_train[idx]
y_train = y_train[idx]
idx = np.argsort(X_val)
X_val = X_val[idx]
y_val = y_val[idx]
```

```python
#The optimum unregularized vs regularized paramemeters
my_alpha = 0.
W0,b0 = GD_reg(X_train,y_train,alpha=my_alpha)
W0 = W0[-1]; b0 = b0[-1]
mse0 = np.mean((W0*X_val+b0-y_val)**2)
print(W0,b0,mse0)
```

```python
# noise generator
def gen_noise(X,noise_level):
    noise = np.random.normal(0,noise_level,len(X))
    return noise
```

```python
noise_level = 0.5
X_noise = X_train + gen_noise(X_train,noise_level)
plt.plot(X_noise,'r')
plt.plot(X_train)
```

# Reg. models are better in noisy test data but not necessarily in clean data

```python
alpha = np.arange(0,1,0.05)
mse_train = []
mse_val = []
mse_noise = []
noise_level = 0.5#noise = 0 reg. doesn't help, but if noise>0 reg. helps
X_noise = X_val+gen_noise(X_val,noise_level) #add noise to validation set
for a in alpha:
    W_,b_ = GD_reg(X_train,y_train,a) #train on clean set
    mse_train.append(np.mean((W_[-1]*X_train+b_[-1]-y_train)**2)) #accumulate error on train set
    mse_val.append(np.mean((W_[-1]*X_val+b_[-1]-y_val)**2)) #accumulate error on validation set
    mse_noise.append(np.mean((W_[-1]*X_noise+b_[-1]-y_val)**2)) #accumulate error on validation set
plt.subplot(122)
plt.plot(alpha, mse_train,'b') #train
plt.plot(alpha, mse_noise,'r') #val_noise  #val_clean: error decreases with reg
plt.plot(alpha, mse_val,'r',linestyle='--') #val_clean: error increases with reg
plt.show()    
```
