SVM Question

```python
import scipy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X= np.array([1,2,3,4,5]).reshape(-1,1)
Y= np.array([-1,-1,-1,1,1])

#show plot, small graph, large circles, different blue and red for labels
plt.scatter(X,Y,s=100, c=Y, cmap=plt.cm.Spectral)
plt.show()
```

```python
#create an svm classifier, and pplot the descision boundary
from sklearn import svm
clf = svm.SVC(kernel='linear')
clf.fit(X,Y)

print(f"'W' Coefficient = {clf.coef_[0][0]} , Intercept (B) = {clf.intercept_[0]}")
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear', C=10000)
clf.fit(X,Y)

print(f"'W' Coefficient = {clf.coef_[0][0]} , Intercept (B) = {clf.intercept_[0]}")
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear', C=0.001)
clf.fit(X,Y)

print(f"'W' Coefficient = {clf.coef_[0][0]} , Intercept (B) = {clf.intercept_[0]}")

clf.predict(X)
```

Linear Regression question (regularization)

```python
X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([1,2,3,4,5])
```

```python
#Fit linear regression
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X, Y)

#Plot the results
plt.scatter(X, Y, color='black', label='Data', s=400, alpha=0.7)
plt.plot(X, regr.predict(X), color='blue', linewidth=3, label='Linear Regression', alpha=0.7)

#plot ridge regression next to it
regr = linear_model.Ridge(alpha=2)
regr.fit(X, Y)
plt.plot(X, regr.predict(X), color='red', linewidth=3, alpha = 0.7, label='Ridge Regression-Alpha=2')
plt.scatter(X, Y, color='black')

regr = linear_model.Ridge(alpha=10)
regr.fit(X, Y)
plt.plot(X, regr.predict(X), color='green', linewidth=3, alpha = 0.7, label='Ridge Regression-Alpha=10')
plt.scatter(X, Y, color='black')

plt.legend(loc='best')

```
