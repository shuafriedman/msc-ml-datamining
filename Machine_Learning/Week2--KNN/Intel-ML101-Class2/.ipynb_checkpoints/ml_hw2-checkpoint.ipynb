```python
# The model is exponential we willtransform to linear via log and then apply linear regression
#X = number of days
#Y = exp(a(X-b)) number of cases is exponential
#a ~ the rate of growth, b ~ the delay in growing


import pandas as pd
import numpy as np
from matplotlib import pyplot
%matplotlib inline
from sklearn.linear_model import LinearRegression
table = pd.read_csv('C:\\WinPython-3.5\\notebooks\\docs\\time_series_2019-ncov-Confirmed.csv').T
```

```python
#country 16 in table is Italy
#country 11 in table is Germany
#country 16 in table is Italy
#country 33 in table is Israel
#country 155 in table is Iran ***linear because of lies***
#country 171 in table is Beijing ****Start decresing?!!!****
country = 155
table[country][4:].plot() 
```

```python
y = table[country][4:].values.astype(float).reshape(-1, 1) #choose country
pyplot.plot(y)
```

```python
#find b, (the first case) in the model manually
b = 40
print(b)
```

```python
X = np.arange(len(y)).reshape(-1, 1)

reg = LinearRegression().fit(X, y)

print(reg.score(X, y))
print(reg.intercept_,reg.coef_)

y_pred = reg.predict(X)
```

```python
#log transform
# we do not fit before b. equivalent to X <- X-b
y_ln = np.log(y+1) #+1 because log(0)
# we do not fit before b
reg_ln = LinearRegression().fit(X[b:], y_ln[b:])

print(reg_ln.score(X[b:], y_ln[b:]))
print(reg_ln.intercept_,reg_ln.coef_)
```

```python

y_ln_pred = y_ln
# back from log transform
y_ln_pred[b:] = reg_ln.predict(X[b:])
y_ln_pred = np.exp(y_ln)
```

```python
pyplot.scatter(X, y_pred, color ='g')
pyplot.scatter(X, y_ln_pred, color ='r')
pyplot.scatter(X, y, color ='b')
```
